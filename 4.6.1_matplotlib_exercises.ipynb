```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
x = np.arange(0,50,5)
plt.plot(x**2 - x + 2,'ro')
plt.annotate('Origin', xy=(0, 0), xytext=(0, 1000), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
#2
x = np.arange(0,10,2)
plt.plot(x**.5,'ro')
# plt.annotate('Intersection', xy=(0, 0), xytext=(0, 1000), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = np.arange(0,10,2)
plt.plot(x**3,'r')
# plt.annotate('Intersection', xy=(0, 0), xytext=(0, 1000), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = np.arange(-10,10,2)
plt.plot(np.tan(x),'--c')
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
x = np.arange(0,10,2)
plt.plot(2**x,'ro')
# plt.annotate('Intersection', xy=(0, 0), xytext=(0, 1000), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
#3
n_rows = 2
n_cols = 2
```

```python
plt.subplot(n_rows, n_cols, 1)
x = np.arange(0,10,2)
plt.plot(x**.5)
plt.title("Square Root")

plt.subplot(n_rows, n_cols, 2)
plt.plot(x**3,'r')
plt.title("Cubed")

plt.subplot(n_rows, n_cols, 3)
plt.plot(np.tan(x),'.g')
plt.title("Tangent")

plt.subplot(n_rows, n_cols, 4)
plt.plot(2**x, '-.k')
plt.title("2^x")
```

```python
#4

x = np.arange(0,10,2)
plt.plot(x**.5, '-y', label='Square Root')
plt.plot(x**3,'ro', label='Cubed')
plt.plot(np.tan(x), label='Tan')
plt.plot(2**x,'-.c', label='x^2')

plt.ylim(0,275)
plt.legend()

plt.show()
```
