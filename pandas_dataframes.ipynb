```python
from pydataset import data
```

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(123)
```

```python
students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']
```

```python
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))
```

```python
df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})
```

```python
df
```

```python
# Create a column named passing_english that indicates whether each student has a passing grade in reading.
df["passing_english"] = df['english'] > 69
```

```python
# Sort the english grades by the passing_english column. How are duplicates handled?
df.sort_values(by="passing_english")
# Duplicates are sorted by the row-index in ascending order
```

```python
# Sort the english grades first by passing_english and then by student name. All the students
# that are failing english should be first, and within the students that are failing english 
# they should be ordered alphabetically. The same should be true for the students passing 
# english. (Hint: you can pass a list to the .sort_values method)
df.sort_values(by=["passing_english","name"])
```

```python
# Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.
df.sort_values(by=["passing_english","english"])
```

```python
df
```

```python
# Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.

#this sums all the values in each row.
#the sum collapses all the fields into one field per row.
#I'll add this as a column
df.sum(axis=1)
```

```python
df['overall_grade'] = df.sum(axis=1)
```

```python
df.sort_values(by='overall_grade')
```

```python
# Rename the cty column to city.
mpg = mpg.rename(columns={"cty": "city"})
mpg.head(10)
```

```python
# Rename the hwy column to highway.
mpg = mpg.rename(columns={"hwy": "highway"})
# I could've done that on the last cell if I knew I was gonna rename two columns. =( 
```

```python
# Do any cars have better city mileage than highway mileage?
mpg[mpg['city'] > mpg['highway']]
#That boolean mask returned no rows. So I'd say no.
```

```python
# Create a column named mileage_difference this column should contain the difference 
# between highway and city mileage for each car.
mpg['mileage_difference'] = mpg['highway'] - mpg['city']
mpg.head(5)
```

```python
# Which car (or cars) has the highest mileage difference?

#Sorted the cars by the indicated column, from greatest to least, and just the top 5
mpg.sort_values(by="mileage_difference", ascending=False).head(5)
```

```python
# Which compact class car has the lowest highway mileage? 
mpg[mpg['class'] == 'compact'].sort_values(by='highway').head(1)
```

```python
# The best?
mpg[mpg['class'] == 'compact'].sort_values(by='highway', ascending=False).head(1)
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg['average_mileage'] = mpg['city'] + mpg['highway'] / 2
mpg.tail(7)
```

```python
# Which dodge car has the best average mileage? 
mpg[mpg['manufacturer'] == 'dodge'].sort_values(by="average_mileage", ascending=False).head(1)
```

```python
# The worst?
mpg[mpg['manufacturer'] == 'dodge'].sort_values(by="average_mileage").head(1)
```
